This notebook demonstrates how to use the Neural Network Efficiency Analyzer to:

## ðŸ“¦ Installation and Setup

```python
# Install required packages (uncomment if needed)
# !pip install tensorflow numpy matplotlib seaborn

import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Import the analyzer
from nn_efficiency import NNEfficiencyAnalyzer, Visualizer, quick_analyze

print("âœ… All packages imported successfully!")
```

## ðŸ”¨ Create a Sample Model

```python
# Create a simple neural network
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(20,), name='hidden_1'),
    keras.layers.Dense(64, activation='relu', name='hidden_2'),
    keras.layers.Dense(32, activation='relu', name='hidden_3'),
    keras.layers.Dense(10, activation='softmax', name='output')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

## ðŸ“Š Generate Training Data and Train Model

```python
# Generate sample data
X_train = np.random.randn(1000, 20)
y_train = np.random.randint(0, 10, 1000)

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

print("\nâœ… Model training complete!")
```

## ðŸ” Quick Analysis (Easy Mode)

```python
# Use the quick_analyze function for one-line analysis
analyzer = quick_analyze(model, X_train[:100], framework='tensorflow', visualize=True)
```

## ðŸ“Š Detailed Analysis (Advanced Mode)

```python
# Create analyzer instance
analyzer = NNEfficiencyAnalyzer(model, framework='tensorflow')

# Perform analysis
results = analyzer.analyze(X_train[:100], compute_activations=True)

# Print detailed report
analyzer.print_report()
```

## ðŸ“ˆ Visualizations

```python
# Create visualizer
viz = Visualizer()

# Plot importance distributions
viz.plot_layer_importance_distribution(analyzer, figsize=(15, 10))
```

```python
# Compare metrics across layers
viz.plot_layer_comparison(analyzer, figsize=(12, 6))
```

```python
# Pruning sensitivity analysis
viz.plot_pruning_sensitivity(analyzer, figsize=(12, 5))
```

```python
# Efficiency radar for first hidden layer
viz.plot_efficiency_radar(analyzer, layer_idx=0, figsize=(8, 8))
```

## ðŸ” Access Detailed Metrics

```python
# Get global metrics
print("Global Metrics:")
for key, value in analyzer.global_metrics.items():
    print(f"  {key}: {value}")

# Get per-layer metrics
print("\nLayer Metrics:")
for la in analyzer.layer_analyses:
    print(f"\n{la.name}:")
    for key, value in la.metrics.items():
        print(f"  {key}: {value:.4f}" if isinstance(value, float) else f"  {key}: {value}")
```

## ðŸ’¡ Optimization Recommendations

Based on the analysis above:

1. **High Redundancy**: If redundancy score > 70%, consider reducing layer width
2. **Low Utilization**: If utilization < 30%, the model may be oversized
3. **High Sparsity**: If sparsity > 50%, the model is a good candidate for pruning
4. **Compression Potential**: Shows how much the model can be compressed
